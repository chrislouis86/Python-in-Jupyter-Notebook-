```python
import numpy as np
import matplotlib.pyplot as plt 

```

```python
y = np.random.randint(0,100,1000)
plt.plot(y)

```

```python
plt.figure(figsize=(10,5))
plt.plot(y)
plt.xlabel('Time')
plt.ylabel('values')


```

```python
y1 = 0.5*np.random.random(1000)

y2 = np.linspace(start = 0, stop = 3,num = 1000)

y3 = y1 + y2

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

```

```python
x = np.linspace(0, 3.2, 1000)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
```
